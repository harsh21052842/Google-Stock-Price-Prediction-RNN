```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```python
# Dataset is entered
dataset_train = pd.read_csv('Google_Stock_Price_Train.csv')
# Particular rows and colums are read
training_set = dataset_train.iloc[:,1:2].values
```

```python
np.set_printoptions(threshold=sys.maxsize,precision=2)
# Printing Training set for test
print(training_set)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create an instance of the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Reshape the training_set if needed
training_set = training_set.reshape(-1, 1)

# Scale the training data
training_set_scaled = scaler.fit_transform(training_set)
```

```python
# In summary, this code snippet creates input sequences (X_train) and 
#  corresponding target values (y_train) by sliding a window over the 
#  training_set_scaled array. These sequences can be used to train a 
#  time series forecasting model, such as a recurrent neural network (RNN) 
#  or a convolutional neural network (CNN), to predict the next value in 
#  the time series based on the previous 60 values.

# Creating a data structure with 60 timesteps and 1 output

X_train = []
y_train = []

for i in range(60, 1258):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])

# Convert the lists to numpy arrays
X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
# We can add some more indicators after reshaping it
X_train = np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1))
#  This reshaping is often required when working with RNNs or CNNs because
#   these models expect input data in a specific shape, typically in the 
#   form of (batch size, sequence length, number of features). In this case, 
#   the reshaping prepares the data to be fed into a model that expects input
#   sequences with a length of 60 and a single feature value at each time step.
X_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
regressor = Sequential()
```

```python
regressor.add(LSTM(units = 50 , return_sequences = True , input_shape = (X_train.shape[1],1)))
regressor.add(Dropout(0.2))
# The line of code regressor.add(Dropout(0.2)) adds a Dropout layer to the neural network 
#   model regressor. Here, 0.2 represents the fraction of input units to drop during
#   training, which is set to 20% in this case.
```

```python
regressor.add(LSTM(units = 50 , return_sequences = True))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units = 50 , return_sequences = True))
regressor.add(Dropout(0.2))
```

```python
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2)) 
```

```python
regressor.add(Dense(units = 1))
```

```python
# We can use RMSprop also by checking which is giving more proper result
regressor.compile(optimizer = 'adam' , loss = 'mean_squared_error')
```

```python
# Fitting the RNN to training set
regressor.fit(X_train, y_train, epochs = 100 , batch_size = 32)
```

```python
dataset_test = pd.read_csv('Google_Stock_Price_Test.csv')
real_stock_price = dataset_test.iloc[:,1:2].values
```

```python
dataset_total = pd.concat((dataset_train['Open'], dataset_test['Open']), axis = 0)
inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values
inputs = inputs.reshape(-1,1)
inputs = scaler.transform(inputs)
X_test = []
for i in range(60, 80):
    X_test.append(inputs[i-60:i, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price = scaler.inverse_transform(predicted_stock_price)
```
